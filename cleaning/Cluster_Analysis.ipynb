# 2. Only Select Top Features and Target Column

```python
df['max_age_difference'] = df['heart_attack_age'] - df['max_age']
```

```python
df_2 = df[['segments', 'heart_attack_age', 'max_age', 'age_chronic_heart_disease', 'angina_pectoris_age',
           'heart_failure_age', 'asthma_age', 'stroke_age', 'heart_attack', 'max_age_difference']]
df_2 = df_2.copy()
df_2.head()
```

# 3. Heart Attacks by Group

### 3.1 Cluster 0

```python
df_cluster_0 = df_2[df_2['segments']==0]
df_cluster_0 = df_cluster_0[['segments', 'heart_attack']]
df_cluster_0 = df_cluster_0.groupby('heart_attack').count().reset_index()
df_cluster_0['segments'].astype(int)
df_cluster_0['percent'] = (df_cluster_0['segments']/3288)*100
df_cluster_0
```

### 3.2 Cluster 1

```python
df_cluster_1 = df_2[df_2['segments']==1]
df_cluster_1 = df_cluster_1[['segments', 'heart_attack']]
df_cluster_1 = df_cluster_1.groupby('heart_attack').count().reset_index()
df_cluster_1['segments'].astype(int)
df_cluster_1['percent'] = (df_cluster_1['segments']/2181)*100
df_cluster_1
```

### 3.2 Cluster 2

```python
df_cluster_2 = df_2[df_2['segments']==2]
df_cluster_2 = df_cluster_2[['segments', 'heart_attack']]
df_cluster_2 = df_cluster_2.groupby('heart_attack').count().reset_index()
df_cluster_2['segments'].astype(int)
df_cluster_2['percent'] = (df_cluster_2['segments']/9)*100
df_cluster_2
```
