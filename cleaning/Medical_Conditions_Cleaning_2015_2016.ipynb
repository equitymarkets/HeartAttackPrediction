# 2. Clean Data

### 2.1 Only Keep Relevant Columns

```python
df_2 = df[['SEQN', 'MCQ025', 'MCQ180B', 'MCQ180C', 'MCQ180D', 'MCQ180E', 'MCQ180F',
           'MCQ180G', 'MCQ180M', 'MCQ180L', 'MCQ300A', 'MCQ300B', 'MCQ300C',
           'MCQ010', 'MCQ160B', 'MCQ160C', 'MCQ160D', 'MCQ160E', 'MCQ160F', 'MCQ160G',
           'MCQ160M', 'MCQ160L']]
df_2 = df_2.copy()
df_2.head()
```

### 2.2 Rename Columns

```python
df_2 = df_2.rename(columns={'SEQN':'id', 'MCQ025':'asthma_age', 'MCQ010':'asthma',
                            'MCQ180B':'heart_failure_age', 'MCQ160B':'heart_failure',
                            'MCQ180C':'coronary_heart_disease_age',
                            'MCQ160C':'coronary_heart_disease',
                            'MCQ180D':'angina_pectoris_age', 'MCQ160D':'angina_pectoris',
                            'MCQ180E':'heart_attack_age', 'MCQ160E':'heart_attack',
                            'MCQ180F':'stroke_age', 'MCQ160F':'stroke',
                            'MCQ180G': 'emphysema_age', 'MCQ160G': 'emphysema',
                            'MCQ180M':'thyroid_age', 'MCQ160M':'thyroid',
                            'MCQ180L':'liver_age', 'MCQ160L':'liver',
                            'MCQ300B':'relative_asthma', 'MCQ300C':'relative_diabetes',
                            'MCQ300A':'relative_heart_attack'})
df_2.head()
```

### 2.3 Create Binary Columns, Fill in Missing Data

```python
df_2['asthma'] = np.where(df_2['asthma']==1, 1, 0)
df_2['heart_failure'] = np.where(df_2['heart_failure']==1, 1, 0)
df_2['coronary_heart_disease'] = np.where(df_2['coronary_heart_disease']==1, 1, 0)
df_2['angina_pectoris'] = np.where(df_2['angina_pectoris']==1, 1, 0)
df_2['heart_attack'] = np.where(df_2['heart_attack']==1, 1, 0)
df_2['stroke'] = np.where(df_2['stroke']==1, 1, 0)
df_2['emphysema'] = np.where(df_2['emphysema']==1, 1, 0)
df_2['thyroid'] = np.where(df_2['thyroid']==1, 1, 0)
df_2['liver'] = np.where(df_2['liver']==1, 1, 0)
df_2['relative_diabetes'] = np.where(df_2['relative_diabetes']==1, 1, 0)
df_2['relative_heart_attack'] = np.where(df_2['relative_heart_attack']==1, 1, 0)
df_2['heart_attack'] = np.where(df_2['heart_attack']==1, 1, 0)
df_2.head()
```

### 2.4 Clean Age Variables

```python
df_2['heart_attack_age'].describe()
```

```python
# 122 highest recorded age source: https://www.cnbc.com/2023/02/21/longevity-expert-3-reasons-the-worlds-oldest-person-lived-to-122.html#:~:text=Jeanne%20Calment%2C%20a%20French%20woman,links%20between%20health%20and%20longevity.
df_3 = df_2.loc[((df_2['heart_attack_age']>=0.0)&(df_2['heart_attack_age']<=122.0))|
                (df_2['heart_attack_age'].isna())]
df_3 = df_3.loc[((df_3['asthma_age']>=0.0)&(df_3['asthma_age']<=122.0))|
                (df_3['asthma_age'].isna())]
df_3 = df_3.loc[((df_3['heart_failure_age']>=0.0)&(df_3['heart_failure_age']<=122.0))|
                (df_3['heart_failure_age'].isna())]
df_3 = df_3.loc[((df_3['coronary_heart_disease_age']>=0.0)&(df_3['coronary_heart_disease_age']<=122.0))|
                (df_3['coronary_heart_disease_age'].isna())]
df_3 = df_3.loc[((df_3['angina_pectoris_age']>=0.0)&(df_3['angina_pectoris_age']<=122.0))|
                (df_3['angina_pectoris_age'].isna())]
df_3 = df_3.loc[((df_3['stroke_age']>=0.0)&(df_3['stroke_age']<=122.0))|
                (df_3['stroke_age'].isna())]
df_3 = df_3.loc[((df_3['thyroid_age']>=0.0)&(df_3['thyroid_age']<=122.0))|
                (df_3['thyroid_age'].isna())]
df_3 = df_3.loc[((df_3['liver_age']>=0.0)&(df_3['liver_age']<=122.0))|
                (df_3['liver_age'].isna())]
df_3 = df_3.loc[((df_3['emphysema_age']>=0.0)&(df_3['emphysema_age']<=122.0))|
                (df_3['emphysema_age'].isna())]
df_3.head()
```

```python
df_4 = df_3.copy()
df_4.info()
```

### 2.5 Calculate Difference in Years between Heart Attack and Comorbidities

```python
df_4['asthma_yrs'] = df_4['heart_attack_age'] - df_4['asthma_age']
df_4['heart_failure_yrs'] = df_4['heart_attack_age'] - df_4['heart_failure_age']
df_4['coronary_heart_disease_yrs'] = df_4['heart_attack_age'] - df_4['coronary_heart_disease_age']
df_4['angina_pectoris_yrs'] = df_4['heart_attack_age'] - df_4['angina_pectoris_age']
df_4['stroke_yrs'] = df_4['heart_attack_age'] - df_4['stroke_age']
df_4['emphysema_yrs'] = df_4['heart_attack_age'] - df_4['emphysema_age']
df_4['thyroid_yrs'] = df_4['heart_attack_age'] - df_4['thyroid_age']
df_4['liver_disease_yrs'] = df_4['heart_attack_age'] - df_4['liver_age']
df_4.head()
```

```python
df_4.describe()
```

```python
df_4[df_4['heart_attack_age'].notna()]
```

### 2.6 Clean Age and Year Difference Columns

```python
yr_columns_to_check = ['asthma_yrs', 'heart_failure_yrs', 'coronary_heart_disease_yrs',
                       'angina_pectoris_yrs', 'stroke_yrs', 'emphysema_yrs',
                       'thyroid_yrs', 'liver_disease_yrs']
```

```python
for column in yr_columns_to_check:
    df_4[column] = np.where((df_4[column].isna())|(df_4[column]<0), 0, df_4[column])
```

```python
age_columns_to_check = ['asthma_age', 'heart_failure_age', 'coronary_heart_disease_age',
                        'angina_pectoris_age', 'stroke_age', 'emphysema_age',
                        'thyroid_age', 'liver_age']
```

```python
for column in age_columns_to_check:
    df_4[column] = np.where((df_4[column].isna())|(df_4[column]>df_4['heart_attack_age']), 0, df_4[column])
```

```python
df_4[df_4['heart_attack_age'].notna()]
```

### 2.6 Add Column for Closest Comorbidity Age

```python
df_4['max_age'] = df_4[age_columns_to_check].max(axis=1)
df_4.head()
```

```python
df_4.to_csv('Resources/medical_conditions_cleaned_2015_2016.csv', index=False)
```
