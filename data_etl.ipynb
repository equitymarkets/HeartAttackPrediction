```python
import pandas as pd
import numpy as np
```

```python
demo_df = pd.read_sas('./Resources/P_DEMO.XPT')
```

```python
demo_df.head()
```

```python
demo_df = demo_df.rename(columns = {'RIDAGEYR': 'Age', 'RIAGENDR': 'Gender','RIDRETH3': 'Race', 'INDFMPIR': 'Income to Poverty Ratio'})
demo_df.head()
```

```python
demo_df_reduced = demo_df[['SEQN', 'Gender', 'Age', 'Race', 'Income to Poverty Ratio']]
demo_df_reduced.head()
```

```python
alc_df = pd.read_sas('./Resources/P_ALQ.XPT')
alc_df.head()
```

```python
alc_df = alc_df.rename(columns={'ALQ121': 'Frequency_of_Drinking', 'ALQ130': 'Drinks_per_Day'})
alc_df_reduced = alc_df[['SEQN', 'Frequency_of_Drinking', 'Drinks_per_Day']]
alc_df_reduced['Frequency_of_Drinking'] = alc_df_reduced['Frequency_of_Drinking'].round()
alc_df_reduced.head()
```

```python
med_conditions_df = pd.read_sas('./Resources/P_MCQ.XPT')
med_conditions_df.head()
```

```python
med_conditions_df = med_conditions_df.rename(columns = {'MCQ160E': 'Heart Attack', 'MCQ160F': 'Stroke'})
med_conditions_df_reduced = med_conditions_df[['SEQN', 'Heart Attack', 'Stroke']]

med_conditions_df_reduced.head()
```

```python
activity_df = pd.read_sas('./Resources/P_PAQ.XPT')
activity_df.head()
```

```python
activity_df = activity_df.rename(columns = {'PAQ605': 'Work', 'PAQ650': 'Recreational'})
activity_df_reduced = activity_df[['SEQN', 'Work', 'Recreational']]
activity_df_reduced = activity_df_reduced.dropna()
activity_df_reduced['Work'] = np.where(activity_df_reduced['Work'] == 1, 1, 0)
activity_df_reduced['Recreational'] = np.where(activity_df_reduced['Recreational'] == 1, 1, 0)
activity_df_reduced.head()
```

```python
smoking_df = pd.read_sas('./Resources/P_SMQ.XPT')
smoking_df.head()
```

```python
smoking_df = smoking_df.rename(columns = {'SMQ020': '100_Cigarettes'})
smoking_df_reduced = smoking_df[['SEQN', '100_Cigarettes']]
smoking_df_reduced = smoking_df_reduced.dropna()
smoking_df_reduced['100_Cigarettes'] = np.where(smoking_df_reduced['100_Cigarettes'] == 1, 1, 0)
smoking_df_reduced.head()
```

```python
diabetes_df = pd.read_sas('./Resources/P_DIQ.XPT')
diabetes_df.head()
```

```python
diabetes_df = diabetes_df.rename(columns={'DIQ010': 'Diabetes'})
diabetes_df_reduced = diabetes_df[['SEQN', 'Diabetes']]
diabetes_df_reduced.head()
```

```python
cholesterol_df = pd.read_sas('./Resources/P_TCHOL.XPT')
cholesterol_df.head()
```

```python
cholesterol_df = cholesterol_df.drop('LBDTCSI', axis=1)
cholesterol_df = cholesterol_df.rename(columns = {'LBXTC': 'Total_Cholesterol_mg_dL'})
cholesterol_df.head()
```

```python
total_df = cholesterol_df.merge(alc_df_reduced, on='SEQN')\
                .merge(activity_df_reduced, on='SEQN')\
                .merge(smoking_df_reduced, on='SEQN')
total_df.head()
```

```python
total_df = total_df.dropna()
total_df = total_df.rename(columns = {'SEQN': 'id'})
total_df.shape
```

```python
total_df.head()
```

```python
total_df.to_csv('./Resources/total_df2.csv', index=False)
```
