```python
import pandas as pd
import numpy as np
```

```python
demo_df = pd.read_sas('./Resources/P_DEMO.XPT', format= 'xport')
demo_df.head()
```

```python
# keeping columns we want 
demo_df = demo_df[['SEQN', 'RIAGENDR', 'RIDAGEYR', 'RIDRETH3']]
demo_df.head()
```

```python
# renaming columns
demo_df = demo_df.rename(columns = {'SEQN': 'id', 'RIAGENDR': 'gender', 'RIDAGEYR': 'age', 'RIDRETH3': 'race'})
demo_df.head()
```

```python
# changing values of columns 
demo_df['gender'] = demo_df['gender'].replace({1: 'male', 2: 'female'})
demo_df['race'] = demo_df['race'].replace({1: 'mexican_american', 2: 'other_hispanic', 3: 'white',
                                           4: 'black', 6: 'asian', 7: 'other'})
demo_df.head()
```

```python
demo_df = pd.get_dummies(demo_df)
demo_df.head()
```

```python
demo_df['gender_female'] = demo_df['gender_female'].replace({True: 1, False: 0})
demo_df['gender_male'] = demo_df['gender_male'].replace({True: 1, False: 0})
demo_df['race_mexican_american'] = demo_df['race_mexican_american'].replace({True: 1, False: 0})
demo_df['race_other_hispanic'] = demo_df['race_other_hispanic'].replace({True: 1, False: 0})
demo_df['race_asian'] = demo_df['race_asian'].replace({True: 1, False: 0})
demo_df['race_black'] = demo_df['race_black'].replace({True: 1, False: 0})
demo_df['race_white'] = demo_df['race_white'].replace({True: 1, False: 0})
demo_df['race_other'] = demo_df['race_other'].replace({True: 1, False: 0})

demo_df.head()
```

```python
demo_df.to_csv('./Resources/demographics.csv', index=False)
```

```python
med_conditions = pd.read_sas('./Resources/P_MCQ.XPT', format = 'xport')
med_conditions.head()
```

```python
age_colmns = ['SEQN', 'MCQ025', 'MCD180B', 'MCD180C', 'MCD180D', 'MCD180E', 'MCD180F',
            'MCD180M', 'MCD180L', 'MCQ570']
med_columns = ['SEQN', 'MCQ010', 'MCQ160B', 'MCQ160C', 'MCQ160D', 'MCQ160F',
           'MCQ160M', 'MCQ160L', 'MCQ560', 'MCQ160E']

```

```python
age_df = med_conditions[age_colmns]
age_df.head()
```

```python
age_df = age_df.rename(columns={'SEQN':'id', 'MCQ025':'asthma_age',
                            'MCD180B':'heart_failure_age',
                            'MCD180C':'age_chronic_heart_disease', 'MCD180D':'angina_pectoris_age',
                            'MCD180E':'heart_attack_age', 'MCD180F':'stroke_age', 'MCD180M':'thyroid_age',
                            'MCD180L':'liver_age', 
                            'MCQ570':'gallbladder_pr_age'})
age_df.head()
```

```python
age_df.shape
```

```python
mc_df = med_conditions[med_columns]
mc_df.head()
```

```python
mc_df = mc_df.rename(columns = {'SEQN': 'id', 'MCQ010': 'asthma', 'MCQ160B': 'heart_failure',
                                 'MCQ160C': 'heart_disease', 'MCQ160D': 'angina_pectoris', 
                                 'MCQ160F': 'stroke', 'MCQ160M': 'thyroid', 'MCQ160L': 'liver_condition', 
                                 'MCQ560': 'gallbladder', 'MCQ160E': 'heart_attack'})
mc_df.head()
```

```python
mc_df = mc_df.dropna()
mc_df.head()
```

```python
mc_df.shape
```

```python
mc_df = mc_df[ (mc_df['asthma']!=9) & (mc_df['heart_failure']!=9) & (mc_df['heart_disease']!=9) & (mc_df['angina_pectoris']!=9) & (mc_df['heart_attack']!=9) & (mc_df['stroke']!=9) & (mc_df['thyroid']!=9) & (mc_df['liver_condition']!=9) & (mc_df['gallbladder']!=9) ]
```

```python
mc_df.shape
```

```python
mc_df['asthma'] = np.where(mc_df['asthma'] == 1, 1, 0)
mc_df['heart_failure'] = np.where(mc_df['heart_failure'] == 1, 1, 0)
mc_df['heart_disease'] = np.where(mc_df['heart_disease'] == 1, 1, 0)
mc_df['angina_pectoris'] = np.where(mc_df['angina_pectoris'] == 1, 1, 0)
mc_df['stroke'] = np.where(mc_df['stroke'] == 1, 1, 0)
mc_df['thyroid'] = np.where(mc_df['thyroid'] == 1, 1, 0)
mc_df['liver_condition'] = np.where(mc_df['liver_condition'] == 1, 1, 0)
mc_df['gallbladder'] = np.where(mc_df['gallbladder'] == 1, 1, 0)
mc_df['heart_attack'] = np.where(mc_df['heart_attack'] == 1, 1, 0)
mc_df.head()
```

```python
df_3 = mc_df.merge(age_df, on = 'id')
df_3.shape
```
