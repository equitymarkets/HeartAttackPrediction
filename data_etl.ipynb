```python
import pandas as pd
import numpy as np
```

```python
demo_df = pd.read_sas('./Resources/P_DEMO.XPT', format= 'xport')
demo_df.head()
```

```python
# keeping columns we want 
demo_df = demo_df[['SEQN', 'RIAGENDR', 'RIDAGEYR', 'RIDRETH3']]
demo_df.head()
```

```python
# renaming columns
demo_df = demo_df.rename(columns = {'SEQN': 'id', 'RIAGENDR': 'gender', 'RIDAGEYR': 'age', 'RIDRETH3': 'race'})
demo_df.head()
```

```python
# changing values of columns 
demo_df['gender'] = demo_df['gender'].replace({1: 'male', 2: 'female'})
demo_df['race'] = demo_df['race'].replace({1: 'mexican_american', 2: 'other_hispanic', 3: 'white',
                                           4: 'black', 6: 'asian', 7: 'other'})
demo_df.head()
```

```python
demo_df = pd.get_dummies(demo_df)
demo_df.head()
```

```python
demo_df['gender_female'] = demo_df['gender_female'].replace({True: 1, False: 0})
demo_df['gender_male'] = demo_df['gender_male'].replace({True: 1, False: 0})
demo_df['race_mexican_american'] = demo_df['race_mexican_american'].replace({True: 1, False: 0})
demo_df['race_other_hispanic'] = demo_df['race_other_hispanic'].replace({True: 1, False: 0})
demo_df['race_asian'] = demo_df['race_asian'].replace({True: 1, False: 0})
demo_df['race_black'] = demo_df['race_black'].replace({True: 1, False: 0})
demo_df['race_white'] = demo_df['race_white'].replace({True: 1, False: 0})
demo_df['race_other'] = demo_df['race_other'].replace({True: 1, False: 0})

demo_df.head()
```

```python
demo_df.to_csv('./Resources/demogrpahics.csv')
```
